```python
import re
def clean_resume_text(text):
    # Remove "Here's a sample/professional resume for X..." prefix
    text = re.sub(r"Here's a (sample |professional )?resume for .+?:\n+", "", text, flags=re.IGNORECASE)
    # Remove closing notes
    text = re.sub(r"Note[:\s].*?(this is just a sample|tailor your resume).*", "", text, flags=re.IGNORECASE|re.DOTALL)
    text = re.sub(r"This is just a sample.*", "", text, flags=re.IGNORECASE|re.DOTALL)
    text = re.sub(r"Remember to.*", "", text, flags=re.IGNORECASE|re.DOTALL)
    text = re.sub(r"Good luck.*", "", text, flags=re.IGNORECASE|re.DOTALL)

    # Remove markdown links [text](url)
    text = re.sub(r'\[([^\]]+)\]\([^\)]+\)', r'\1', text)

    # Remove email
    text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', '', text)
    # Remove phone numbers
    text = re.sub(r'(\+?\(?\d[\d\s\-().]{6,}\d)', '', text)
    # Remove URLs
    text = re.sub(r'linkedin\.com/in/\S+', '', text, flags=re.IGNORECASE)
    text = re.sub(r'github\.com/\S+', '', text, flags=re.IGNORECASE)
    text = re.sub(r'http\S+|www\.\S+', '', text, flags=re.IGNORECASE)

    # Remove Contact Information section
    text = re.sub(r'Contact Information:?', '', text, flags=re.IGNORECASE)

    # ── NEW: Remove leftover labels with empty values ──
    text = re.sub(r'\*?\s*(Email|Phone|Mobile|LinkedIn|GitHub|Address|Location)\s*:?\s*\*?', '', text, flags=re.IGNORECASE)

    # Remove empty bullet points
    text = re.sub(r'\*\s*:\s*', '', text)
    text = re.sub(r'\*\s*\*', '', text)
    text = re.sub(r'\*\s+\*', '', text)

    # Remove name + role title (first non-empty short lines)
    lines = [l.strip() for l in text.split('\n') if l.strip()]
    while lines and re.match(r'^[A-Za-z\s]{2,40}$', lines[0]) and len(lines[0].split()) <= 4:
        lines = lines[1:]
    text = ' '.join(lines)

    # Final cleanup
    text = re.sub(r'\n+', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply
filtered_df['Resume_clean'] = filtered_df['Resume'].apply(clean_resume_text)

# Verify 3 different resumes
for i in [0, 1, 2]:
    print(f"\n--- Resume {i} ---")
    print(filtered_df['Resume_clean'].iloc[i][:400])
    print()
```
