```python
import pandas as pd
from bs4 import BeautifulSoup

df = pd.read_csv("Resume.csv")

# Clean all resumes
df['Resume_clean'] = df['Resume_html'].apply(
    lambda x: BeautifulSoup(x, 'html.parser').get_text(separator=' ', strip=True)
)

# See ONE full resume clearly
print(df['Resume_clean'][0])
```
